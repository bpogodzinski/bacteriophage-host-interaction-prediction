```python
import pandas as pd
import plotnine as p9
import numpy as np
import json
from random import choice
from pathlib import Path

filename = '/home/panda/workspace/wirusy/test/classifiers-test-2022-04-13-T-02-41-57.csv'
dataframe = pd.read_csv(filename, header=0, index_col=[0, 1]).sort_values(by='prediction', ascending=False)
```

```python
correct_predictions = {
                'classificator': [],
                'sampling': [],
                'superkingdom': [],
                'phylum': [],
                'class': [],
                'order': [],
                'family': [],
                'genus': [],
                'species': []
                }
virus_json = json.loads(Path('/home/panda/workspace/wirusy/virus.json').read_text())
host_json = json.loads(Path('/home/panda/workspace/wirusy/host.json').read_text())

for classificator, df_classificator in dataframe.groupby('classificator'):
    for sampling, df_sampling in df_classificator.groupby('sampling'):
        superkingdom = 0
        phylum = 0
        _class = 0
        order = 0
        family = 0
        genus = 0
        species = 0
        for virus, df_virus in df_sampling.groupby(level='virus'):
            max_scores = df_virus[df_virus == df_virus.max()].dropna()
            hosts = max_scores.index.get_level_values('host')
            random_host = choice(hosts)
            results = [x == y for x,y in zip(virus_json[virus]['host']['lineage_names'], host_json[random_host]['lineage_names'])]
            if results[0]:
                superkingdom += 1
            if results[1]:
                phylum += 1
            if results[2]:
               _class += 1
            if results[3]:
                order += 1
            if results[4]:
                family += 1
            if results[5]:
                genus += 1
            if results[6]:
                species += 1
        max_value = max([superkingdom, phylum, _class, order, family, genus, species])
        correct_predictions['classificator'].append(classificator)
        correct_predictions['sampling'].append(sampling)
        correct_predictions['superkingdom'].append(superkingdom / max_value)
        correct_predictions['phylum'].append(phylum / max_value)
        correct_predictions['class'].append(_class / max_value)
        correct_predictions['order'].append(order / max_value)
        correct_predictions['family'].append(family / max_value)
        correct_predictions['genus'].append(genus / max_value)
        correct_predictions['species'].append(species / max_value)

results_df = pd.DataFrame(correct_predictions)
results_df
# p9.ggplot(data=results_df, mapping=p9.aes(x=, y='prediction', label='percent')) \
#     + p9.geom_bar(stat='identity') \
#     + p9.facet_wrap('~ classificator + sampling') \
#     + p9.scale_y_continuous(breaks=np.arange(0, 1.1, 0.1), limits=[0, 1]) \
#     + p9.labs(title='Correct host predictions to taxonomic groups', y='Frequency', x='Taxonomic group') \
#     + p9.geom_text(position=p9.position_stack(vjust=0.9)) \
#     + p9.theme_seaborn() \
#     + p9.theme(figure_size=(15,15),
#                 panel_grid=p9.element_line(color='darkgrey'),
#                 panel_grid_major=p9.element_line(size=1.4, alpha=1),
#                 panel_grid_major_x=p9.element_line(
#                     linetype='dashed'),
#                 panel_grid_major_y=p9.element_line(
#                     linetype='dashdot'),
#                 panel_grid_minor=p9.element_line(alpha=.25),
#                 panel_grid_minor_x=p9.element_line(color='grey'),
#                 panel_grid_minor_y=p9.element_line(color='grey')) 
```
